O método `squeeze` do pandas é utilizado para converter um DataFrame ou uma Série de um objeto que contém apenas uma dimensão (ou seja, uma coluna ou uma linha única) em um objeto mais simples (ou seja, uma Série ou um escalar). 

Aqui estão os pontos principais sobre o método `squeeze`:

### Aplicações do `squeeze`

1. **Transformar DataFrame de uma coluna em Série:**
   Quando você tem um DataFrame com apenas uma coluna, `squeeze` pode ser utilizado para transformar esse DataFrame em uma Série.
   
2. **Transformar DataFrame de uma linha em Série:**
   Similarmente, quando você tem um DataFrame com apenas uma linha, `squeeze` pode ser utilizado para transformar esse DataFrame em uma Série.
   
3. **Transformar Série com um único valor em escalar:**
   Quando você tem uma Série com apenas um valor, `squeeze` pode converter essa Série em um valor escalar (por exemplo, um número ou uma string).

### Sintaxe

```python
DataFrame.squeeze(axis=None)
Series.squeeze(axis=None)
```

### Parâmetros

- `axis`: Pode ser `None`, `0` ou `1`.
  - `None`: Remove todas as dimensões de comprimento 1.
  - `0`: Remove o comprimento 1 das linhas.
  - `1`: Remove o comprimento 1 das colunas.

### Exemplos

#### Exemplo 1: DataFrame com uma única coluna

```python
import pandas as pd

# DataFrame com uma única coluna
df = pd.DataFrame({'A': [1, 2, 3]})
print("DataFrame original:\n", df)

# Usando squeeze para transformar em uma Série
squeezed = df.squeeze()
print("\nDepois do squeeze:\n", squeezed)
```

#### Exemplo 2: DataFrame com uma única linha

```python
# DataFrame com uma única linha
df = pd.DataFrame({'A': [1], 'B': [2], 'C': [3]})
print("DataFrame original:\n", df)

# Usando squeeze para transformar em uma Série
squeezed = df.squeeze()
print("\nDepois do squeeze:\n", squeezed)
```

#### Exemplo 3: Série com um único valor

```python
# Série com um único valor
s = pd.Series([42])
print("Série original:\n", s)

# Usando squeeze para transformar em um escalar
squeezed = s.squeeze()
print("\nDepois do squeeze:\n", squeezed)
```

### Considerações

- O método `squeeze` é particularmente útil quando você trabalha com subconjuntos de dados e deseja simplificar a estrutura dos dados para análises ou operações subsequentes.
- Quando `squeeze` é aplicado a um DataFrame ou Série que não atende aos critérios de transformação (por exemplo, um DataFrame com mais de uma coluna), ele simplesmente retorna o objeto original sem modificações.

O `squeeze` é uma ferramenta prática do pandas para reduzir a dimensionalidade de objetos quando apropriado, facilitando o trabalho com dados em diferentes formas e tamanhos.

No pandas, o código `alcohol.size == alcohol.shape[0]` verifica se o número total de elementos na série ou DataFrame `alcohol` é igual ao número de linhas da mesma.

Aqui está o que cada parte do código faz:

- `alcohol.size`: Retorna o número total de elementos (ou entradas) na série ou DataFrame. Para uma série, isso é o número total de valores. Para um DataFrame, é o número total de células (linhas * colunas).
- `alcohol.shape[0]`: Retorna o número de linhas na série ou DataFrame.

Ao comparar `alcohol.size` com `alcohol.shape[0]`, estamos verificando se o número total de elementos é igual ao número de linhas. 

Essa comparação será verdadeira (True) quando `alcohol` for uma série, pois em uma série, o número de elementos (`size`) será igual ao número de linhas (`shape[0]`). Em um DataFrame, isso geralmente será falso (False), exceto em casos especiais, como quando o DataFrame tem exatamente uma coluna.

Portanto, o código basicamente verifica se `alcohol` é uma série ou um DataFrame de uma única coluna.

## 36. Unique Values And Series Monotonicity

### `unique`

- **Função**: O método `unique` retorna uma array com todos os valores exclusivos de uma série.
- **Uso comum**: Identificar quais valores distintos estão presentes em uma série.
- **Exemplo**:

```python
import pandas as pd

# Criando uma Série de exemplo
data = pd.Series([1, 2, 2, 3, 4, 4, 4, 5])

# Usando unique
valores_unicos = data.unique()
print(valores_unicos)
```

Saída:
```
[1 2 3 4 5]
```

### `nunique`

- **Função**: O método `nunique` retorna o número de valores exclusivos em uma série ou DataFrame.
- **Uso comum**: Contar quantos valores distintos existem em uma série ou em cada coluna de um DataFrame.
- **Exemplo**:

```python
import pandas as pd

# Criando uma Série de exemplo
data = pd.Series([1, 2, 2, 3, 4, 4, 4, 5])

# Usando nunique
numero_valores_unicos = data.nunique()
print(numero_valores_unicos)
```

Saída:
```
5
```

### Resumo das Diferenças

- `unique`: Retorna os valores distintos presentes na série. Útil para ver quais são esses valores.
- `nunique`: Retorna a contagem dos valores distintos presentes na série. Útil para saber quantos valores distintos existem.

### Uso em DataFrames

- **`unique`** não é um método diretamente aplicável a um DataFrame, mas pode ser usado em conjunto com a aplicação de funções em colunas específicas.
- **`nunique`** pode ser aplicado diretamente a um DataFrame para contar valores únicos por coluna:

```python
import pandas as pd

# Criando um DataFrame de exemplo
data = pd.DataFrame({
    'A': [1, 2, 2, 3],
    'B': [1, 1, 2, 2],
    'C': [3, 3, 3, 3]
})

# Usando nunique em um DataFrame
valores_unicos_por_coluna = data.nunique()
print(valores_unicos_por_coluna)
```

Saída:
```
A    3
B    2
C    1
dtype: int64
```

O método `is_monotonic` no pandas é usado para verificar se os valores de uma série ou de uma coluna em um DataFrame são monotônicos. Isso significa que os valores são sempre crescentes ou sempre decrescentes ao longo do índice. Existem variantes deste método, como `is_monotonic_increasing` e `is_monotonic_decreasing`, que verificam especificamente se os valores são estritamente crescentes ou estritamente decrescentes, respectivamente.

### `is_monotonic`

- **Função**: Verifica se os valores de uma série são monotônicos, ou seja, se nunca diminuem ao longo do índice (são sempre crescentes ou constantes).
- **Uso comum**: Verificar a ordenação dos dados.

### Exemplos

#### Verificando se uma série é monotônica

```python
import pandas as pd

# Série de exemplo (monotonicamente crescente)
data1 = pd.Series([1, 2, 2, 4, 5])

# Série de exemplo (não monotônica)
data2 = pd.Series([1, 3, 2, 4, 5])

# Verificando se são monotônicas
print(data1.is_monotonic)  # True
print(data2.is_monotonic)  # False
```

#### `is_monotonic_increasing`

- **Função**: Verifica se os valores de uma série são estritamente crescentes ou constantes.
- **Exemplo**:

```python
print(data1.is_monotonic_increasing)  # True
print(data2.is_monotonic_increasing)  # False
```

#### `is_monotonic_decreasing`

- **Função**: Verifica se os valores de uma série são estritamente decrescentes ou constantes.
- **Exemplo**:

```python
# Série de exemplo (monotonicamente decrescente)
data3 = pd.Series([5, 4, 4, 2, 1])

# Série de exemplo (não monotônica)
data4 = pd.Series([5, 4, 3, 2, 4])

print(data3.is_monotonic_decreasing)  # True
print(data4.is_monotonic_decreasing)  # False
```

### Resumo

- **`is_monotonic`**: Verifica se os valores são sempre crescentes ou constantes.
- **`is_monotonic_increasing`**: Verifica se os valores são estritamente crescentes ou constantes.
- **`is_monotonic_decreasing`**: Verifica se os valores são estritamente decrescentes ou constantes.

Esses métodos são úteis para analisar a tendência e a ordenação dos dados em uma série ou coluna de um DataFrame no pandas.

No pandas, `count()` e `size` são métodos usados para contar elementos, mas eles têm comportamentos diferentes. Aqui está a diferença entre eles:

1. **`count()`**:
   - **Descrição**: Conta o número de entradas não-nulas (não-NaN) em cada coluna ou linha.
   - **Uso**: Pode ser usado em DataFrames e Series.
   - **Sintaxe**: `df.count()` ou `series.count()`
   - **Exemplo**:
     ```python
     import pandas as pd
     data = {'A': [1, 2, None], 'B': [4, None, None]}
     df = pd.DataFrame(data)
     print(df.count())
     ```
     **Saída**:
     ```
     A    2
     B    1
     dtype: int64
     ```
     Aqui, `count()` retorna o número de valores não-nulos em cada coluna.

2. **`size`**:
   - **Descrição**: Retorna o número total de elementos no DataFrame ou Series. Isso inclui valores nulos (NaN).
   - **Uso**: Pode ser usado em DataFrames e Series.
   - **Sintaxe**: `df.size` ou `series.size`
   - **Exemplo**:
     ```python
     import pandas as pd
     data = {'A': [1, 2, None], 'B': [4, None, None]}
     df = pd.DataFrame(data)
     print(df.size)
     ```
     **Saída**:
     ```
     6
     ```
     Aqui, `size` retorna o número total de elementos no DataFrame, que é 6 (3 linhas * 2 colunas).

### Resumo:

- `count()`: Conta o número de valores não-nulos.
- `size`: Conta o número total de elementos, incluindo valores nulos.

Essa distinção é importante dependendo do que você precisa medir em seus dados.

O código `alcohol[alcohol.isnull()]` é usado para filtrar e selecionar elementos nulos (NaN) de uma Série ou coluna específica de um DataFrame no pandas. Vamos detalhar passo a passo:

1. **`alcohol.isnull()`**:
   - Este método retorna uma Série booleana do mesmo tamanho que `alcohol`, onde cada valor é `True` se o elemento correspondente em `alcohol` for nulo (NaN) e `False` caso contrário.

2. **`alcohol[alcohol.isnull()]`**:
   - Este é um processo de indexação, onde você está usando a Série booleana gerada por `alcohol.isnull()` para selecionar apenas os elementos de `alcohol` que são `True` (ou seja, onde os valores são nulos).

Vamos ver um exemplo prático:

```python
import pandas as pd
import numpy as np

# Criando uma Série com alguns valores nulos
alcohol = pd.Series([0.8, 1.2, np.nan, 0.5, np.nan, 0.7])

# Selecionando os elementos nulos
nulos = alcohol[alcohol.isnull()]

print(nulos)
```

**Saída**:
```
2   NaN
4   NaN
dtype: float64
```

Nesse exemplo:
- `alcohol.isnull()` gera uma Série booleana: `[False, False, True, False, True, False]`.
- `alcohol[alcohol.isnull()]` usa essa Série booleana para selecionar os elementos na posição 2 e 4, que são nulos.

Portanto, o código `alcohol[alcohol.isnull()]` filtra e retorna todos os elementos nulos da Série `alcohol`.

````python
alcohol[alcohol.isnull()].index
````

````python
type(alcohol[alcohol.isnull()].index)
````

Vamos analisar os dois códigos para entender a diferença e por que o segundo é mais "adorável":

1. **Código 1:**
   ```python
   len(list(alcohol[alcohol.isnull()].index))
   ```
   - **Explicação**:
     - `alcohol.isnull()`: Cria uma Série booleana indicando onde os valores são nulos (True para nulo, False para não nulo).
     - `alcohol[alcohol.isnull()]`: Filtra a Série original, mantendo apenas os valores nulos.
     - `.index`: Obtém os índices desses valores nulos.
     - `list(...)`: Converte o Índice obtido em uma lista.
     - `len(...)`: Calcula o comprimento da lista, ou seja, o número de elementos nulos.

2. **Código 2:**
   ```python
   alcohol.isnull().sum()
   ```
   - **Explicação**:
     - `alcohol.isnull()`: Cria uma Série booleana indicando onde os valores são nulos (True para nulo, False para não nulo).
     - `.sum()`: Soma os valores booleanos, tratando `True` como 1 e `False` como 0. Portanto, o resultado é a contagem de valores nulos.

### Comparação

1. **Eficiência e Clareza:**
   - **Primeiro Código**:
     - Envolve múltiplas operações: filtragem, extração de índice, conversão para lista e cálculo do comprimento. Cada uma dessas operações tem um custo computacional.
     - Mais verboso e menos direto, tornando o código mais difícil de entender rapidamente.

   - **Segundo Código**:
     - Simplesmente soma os valores booleanos. É uma operação direta e eficiente.
     - Claro e conciso, expressa a intenção diretamente: contar valores nulos.

### Por que o segundo é mais "adorável"?

1. **Desempenho**:
   - O segundo código é mais eficiente em termos de desempenho. Realiza a contagem diretamente sem precisar de operações intermediárias desnecessárias.

2. **Legibilidade**:
   - O segundo código é mais legível e fácil de entender. A intenção do código é clara à primeira vista: contar valores nulos.

3. **Simplicidade**:
   - Menos código e operações redundantes tornam o segundo método mais simples e elegante.

### Exemplo Prático:

```python
import pandas as pd
import numpy as np

alcohol = pd.Series([0.8, 1.2, np.nan, 0.5, np.nan, 0.7])

# Primeiro método
count_nulls_1 = len(list(alcohol[alcohol.isnull()].index))

# Segundo método
count_nulls_2 = alcohol.isnull().sum()

print(count_nulls_1)  # Saída: 2
print(count_nulls_2)  # Saída: 2
```

Ambos os métodos retornam o mesmo resultado (2), mas o segundo método (`alcohol.isnull().sum()`) é preferível devido à sua eficiência, legibilidade e simplicidade.

## 39. BONUS: Another Approach

No contexto do pandas, a vetorização refere-se à execução de operações em toda a estrutura de dados (como DataFrames e Series) de uma vez, em vez de usar loops explícitos para processar elementos individuais. Isso é possível porque <span style="color:lightgreen">pandas é construído em cima do NumPy, que suporta operações vetorizadas eficientes</span>.

### Exemplo:

Vamos considerar um DataFrame que contém dados de vendas de uma loja.

```python
import pandas as pd

# Criando um DataFrame exemplo
dados = {
    'produto': ['A', 'B', 'C'],
    'preco_unitario': [10, 20, 30],
    'quantidade_vendida': [100, 150, 200]
}

df = pd.DataFrame(dados)
print(df)
```

Saída:
```
  produto  preco_unitario  quantidade_vendida
0       A              10                100
1       B              20                150
2       C              30                200
```

### Sem Vetorização (usando loop):

Se você quiser calcular a receita total para cada produto (preço unitário * quantidade vendida), você pode fazer isso usando um loop:

```python
receitas = []

for i in range(len(df)):
    receita = df['preco_unitario'][i] * df['quantidade_vendida'][i]
    receitas.append(receita)

df['receita_total'] = receitas
print(df)
```

Saída:
```
  produto  preco_unitario  quantidade_vendida  receita_total
0       A              10                100           1000
1       B              20                150           3000
2       C              30                200           6000
```

### Com Vetorização:

Usando vetorização, você pode calcular a receita total sem usar um loop explícito:

```python
df['receita_total'] = df['preco_unitario'] * df['quantidade_vendida']
print(df)
```

Saída:
```
  produto  preco_unitario  quantidade_vendida  receita_total
0       A              10                100           1000
1       B              20                150           3000
2       C              30                200           6000
```

### Benefícios da Vetorização:

1. **Performance:** As operações vetorizadas são geralmente mais rápidas porque são implementadas em C e otimizadas para desempenho.
2. **Código Mais Limpo:** Evita a necessidade de escrever loops explícitos, resultando em código mais conciso e legível.
3. **Uso de Memória:** As operações vetorizadas podem ser mais eficientes em termos de memória, pois evitam a criação de listas temporárias.

### Outro Exemplo:

Considere que você deseja aplicar um desconto de 10% em todos os preços unitários:

```python
df['preco_unitario_com_desconto'] = df['preco_unitario'] * 0.9
print(df)
```

Saída:
```
  produto  preco_unitario  quantidade_vendida  receita_total  preco_unitario_com_desconto
0       A              10                100           1000                        9.0
1       B              20                150           3000                       18.0
2       C              30                200           6000                       27.0
```

Com a vetorização, a aplicação do desconto é feita em uma única linha de código, aplicando a operação a toda a coluna de uma vez.

A vetorização, portanto, facilita o trabalho com grandes conjuntos de dados de maneira eficiente e elegante.
